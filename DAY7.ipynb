```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import files
files.download('day6_gacha_costs.xlsx')

# Ê®°ÂûãÂèÉÊï∏
p = 0.005           # ‰∏≠SSRÊ©üÁéá
pity_limit = 100      # ÈªûÊï∏‰øùÂ∫ï‰∏äÈôê
cost_per_draw = 50   # ÊØèÊäΩÊàêÊú¨ÔºàÂÖÉÔºâ
trials = 10000       # Ê®°Êì¨Ê¨°Êï∏

# Áµ±Ë®àÁî®ËÆäÊï∏
cost_list = []
by_lucky = 0
by_pity = 0

for _ in range(trials):
    points = 0
    for draw in range(1, 1000):
        points += 1
        if random.random() < p:
            cost_list.append(draw * cost_per_draw)
            by_lucky += 1
            break
        elif points >= pity_limit:
            cost_list.append(draw * cost_per_draw)
            by_pity += 1
            break

# Áµ±Ë®àÁµêÊûú
avg_cost = np.mean(cost_list)
median_cost = np.median(cost_list)
std_cost = np.std(cost_list)

print(f"üéØ Âπ≥ÂùáÊàêÊú¨ÂèñÂæóUPËßíËâ≤Ôºö${avg_cost:.2f} ÂÖÉ")
print(f"üìä ‰∏≠‰ΩçÊï∏ÊàêÊú¨Ôºö${median_cost:.2f} ÂÖÉ")
print(f"‚öñ Ê®ôÊ∫ñÂ∑ÆÔºö${std_cost:.2f} ÂÖÉ")
print(f"üîπ ÈÅãÊ∞£ÊäΩ‰∏≠Ôºö{by_lucky} Ê¨°Ôºà{by_lucky/trials*100:.1f}%Ôºâ")
print(f"üî∏ ‰øùÂ∫ïÂÖåÊèõÔºö{by_pity} Ê¨°Ôºà{by_pity/trials*100:.1f}%Ôºâ")

# Áï´ÊàêÊú¨ÂàÜÂ∏ÉÁõ¥ÊñπÂúñ
max_cost = max(cost_list)
bins = range(0, int(max_cost) + cost_per_draw, cost_per_draw*4)

plt.figure(figsize=(12,6))
plt.hist(cost_list, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label="Ëä±Ë≤ªÂàÜÂ∏É")

# Áï´Á¥ØÁ©çÂàÜÂ∏ÉÂáΩÊï∏ (CDF)
sorted_costs = np.sort(cost_list)
cdf = np.arange(1, len(sorted_costs)+1) / len(sorted_costs)
plt.plot(sorted_costs, cdf * max(np.histogram(cost_list, bins=bins)[0]), color='red', label="Á¥ØÁ©çÂàÜÂ∏ÉÂáΩÊï∏ (CDF)")

# Ê®ôÁ§∫Âπ≥ÂùáËàá‰∏≠‰ΩçÊï∏
plt.axvline(avg_cost, color='green', linestyle='--', label=f"Âπ≥ÂùáËä±Ë≤ª = ${avg_cost:.2f}")
plt.axvline(median_cost, color='orange', linestyle='--', label=f"‰∏≠‰ΩçÊï∏ = ${median_cost:.2f}")

plt.title("Day 6ÔºöË™≤ÈáëÊàêÊú¨Ê®°Êì¨ÔºàÂê´‰øùÂ∫ïÈªûÊï∏Ôºâ")
plt.xlabel("Á∏ΩËä±Ë≤ª (ÂÖÉ)")
plt.ylabel("Áé©ÂÆ∂‰∫∫Êï∏")
plt.legend()
plt.grid(True)
plt.show()

# ÂåØÂá∫ÁµêÊûúÊàêExcelÊ™î
df = pd.DataFrame({'Cost': cost_list})
df.to_excel('day6_gacha_costs.xlsx', index=False)
print("ExcelÊ™î day6_gacha_costs.xlsx Â∑≤ÂÑ≤Â≠òÂÆåÊàê")
```
