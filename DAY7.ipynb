```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import files
files.download('day6_gacha_costs.xlsx')

# 模型參數
p = 0.005           # 中SSR機率
pity_limit = 100      # 點數保底上限
cost_per_draw = 50   # 每抽成本（元）
trials = 10000       # 模擬次數

# 統計用變數
cost_list = []
by_lucky = 0
by_pity = 0

for _ in range(trials):
    points = 0
    for draw in range(1, 1000):
        points += 1
        if random.random() < p:
            cost_list.append(draw * cost_per_draw)
            by_lucky += 1
            break
        elif points >= pity_limit:
            cost_list.append(draw * cost_per_draw)
            by_pity += 1
            break

# 統計結果
avg_cost = np.mean(cost_list)
median_cost = np.median(cost_list)
std_cost = np.std(cost_list)

print(f"🎯 平均成本取得UP角色：${avg_cost:.2f} 元")
print(f"📊 中位數成本：${median_cost:.2f} 元")
print(f"⚖ 標準差：${std_cost:.2f} 元")
print(f"🔹 運氣抽中：{by_lucky} 次（{by_lucky/trials*100:.1f}%）")
print(f"🔸 保底兌換：{by_pity} 次（{by_pity/trials*100:.1f}%）")

# 畫成本分布直方圖
max_cost = max(cost_list)
bins = range(0, int(max_cost) + cost_per_draw, cost_per_draw*4)

plt.figure(figsize=(12,6))
plt.hist(cost_list, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label="花費分布")

# 畫累積分布函數 (CDF)
sorted_costs = np.sort(cost_list)
cdf = np.arange(1, len(sorted_costs)+1) / len(sorted_costs)
plt.plot(sorted_costs, cdf * max(np.histogram(cost_list, bins=bins)[0]), color='red', label="累積分布函數 (CDF)")

# 標示平均與中位數
plt.axvline(avg_cost, color='green', linestyle='--', label=f"平均花費 = ${avg_cost:.2f}")
plt.axvline(median_cost, color='orange', linestyle='--', label=f"中位數 = ${median_cost:.2f}")

plt.title("Day 6：課金成本模擬（含保底點數）")
plt.xlabel("總花費 (元)")
plt.ylabel("玩家人數")
plt.legend()
plt.grid(True)
plt.show()

# 匯出結果成Excel檔
df = pd.DataFrame({'Cost': cost_list})
df.to_excel('day6_gacha_costs.xlsx', index=False)
print("Excel檔 day6_gacha_costs.xlsx 已儲存完成")
```
